# Advanced graphs

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot('Occupation',hue='Gender', data=data)
```

```python
#plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
sns.countplot('Product_Category',hue='Gender' , data=data)
```

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot('Occupation', hue='Age Group'  , data=data)
```

```python
#plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot('Occupation', hue='Age Group'  , data=data, edgecolor = sns.color_palette('dark', 3))
```

# Purchase power analysis
